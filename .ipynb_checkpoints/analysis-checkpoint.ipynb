# Data Overview
The data we use for this analysis comes from Messari, but we've stored it in CSV files so that you can easily re-run the code and replicate the analysis. We have this data for each token listed in the `token_asset` list.

## 1. **Basic data**
For each token in our dataset, we have the following key basic variables:

* **`timestamp`**: The date of each observation.
* **`symbol`**: The token ticker.
* **`unlockNative`**: The number of tokens unlocked at a given timestamp.
* **`unlock (% of CS)`**: The percentage of the circulating supply unlocked at a given timestamp.
* **`CS`**: The circulating supply of the token at the given timestamp.
* **`TS`**: The total supply of the token.
* **`days since TGE`**: The number of days since the Token Generation Event (TGE).
* **`close`**: The daily closing price of the token.
* **`BTC_close`**: The daily closing price of Bitcoin.

## 2. **Feature Engineering**
To better understand how token unlocks impact price movements, we derive several additional variables from existing ones. 

### Circulating supply variations
To estimate the effect of unlock events on price, we must control from potential impacts of adjacent unlocks on the price, otherwise we may bias our results. We calculate the percentage change in circulating supply over various time windows, before and after the unlock event:
   - **`CS variation 7 past days`**, **`CS variation 8-14 past days`**, etc.: These calculate how much the circulating supply has changed over different periods in the past.
   - **Following Days Inflation**: This tracks changes in circulating supply occuring in the periods following the unlock event (e.g., **`CS variation 8-14 following days`**, etc.).

### Price Variation
To measure the impact of unlocks on price, we calculate daily and multi-day price variations:
   - **`price variation unlock day`**: The daily percentage change in token price on the unlock day.
   - **Log Variations**: We use lagged and leading log price changes to observe price variations over days before and after the unlock (e.g., **`price variation D-1`**, **`price variation D+1`**).
   - **Multi-day Variations**: We calculate price changes over 3-day and 7-day periods before and after each unlock event (e.g., **`price variation following 3 days`**, **`price variation following 4-6 days`**).

### Market Conditions
To control for general market trends, we include several features that track Bitcoin price variations over time:
   - **`BTC price variation past 7 days`** and **`BTC price variation following 7 days`**: Capture how Bitcoinâ€™s price varies over weekly intervals.
   - **Monthly Bitcoin Variation**: This includes monthly past and future variations.

### Daily Unlocks
We create a dummy variable (**`daily unlock`**) that identifies daily unlocks. These are defined as unlock events that are close in size to the unlock amounts on the two adjacent timestamp, for a given symbol. This allows us to isolate the specific types of unlocks (see Part *2: Daily Unlocks* for more details).

### Date filter
We include a filter for unlocks after **2022-01-01**, as the data before this date is of lower quality. Earlier data covers only a few projects and unlocks, giving too much weight to the projects that have been active for longer. By focusing on unlocks after 2022, we ensure a reasonable number of projects, resulting in a significant number of observations and more diverse data within each period. 



```python
final_dataset = None

price_and_volumes_ETH = pd.read_csv(f'data/price and volume/ETH.csv')
price_and_volumes_BTC = pd.read_csv(f'data/price and volume/BTC.csv')
        
for symbol in asset_list['symbol']:
        token_unlocks = pd.read_csv(f'data/token unlocks/{symbol}.csv')
        price_and_volumes = pd.read_csv(f'data//price and volume/{symbol}.csv')


        
        token_data = merge_on_timestamp([token_unlocks,price_and_volumes,
                                            price_and_volumes_ETH,price_and_volumes_BTC]).sort_values(by='timestamp')
        
        
        token_data=token_data[['timestamp', 'unlockNative', 'unlock%CS', 'CS', 'TS',
           'days since TGE', 'open', 'high',
           'low', 'close', 'volume', 'ETH_close',
           'ETH_open', 'ETH_volatility', 'ETH_variation',
           'BTC_close', 'BTC_open', 'BTC_volatility', 'BTC_variation']]
            
        token_data.rename(columns={'unlock%CS': 'unlock (% of CS)'}, inplace=True)
        
        token_data['symbol'] = symbol
        
        token_data['daily unlock'] = (
            (0.9 * token_data['unlockNative'].copy().shift(1) < token_data['unlockNative']) & 
            (token_data['unlockNative'] < 1.1 * token_data['unlockNative'].copy().shift(1)) & 
            (0.9 * token_data['unlockNative'].copy().shift(-1) < token_data['unlockNative']) & 
            (token_data['unlockNative'] < 1.1 * token_data['unlockNative'].copy().shift(-1)) &
            (token_data['unlockNative'] > 0)
        )
        
        #Control variables adjacent unlocks
        token_data['CS variation 7 past days']=token_data['CS'].copy().diff(7).shift(1)/token_data['CS'].copy().shift(8)
        token_data['CS variation 8-14 past days']=token_data['CS'].copy().diff(7).shift(8)/token_data['CS'].copy().shift(15)
        token_data['CS variation 15-21 past days']=token_data['CS'].copy().diff(7).shift(15)/token_data['CS'].copy().shift(22)
        token_data['CS variation 22-28 past days']=token_data['CS'].copy().diff(7).shift(22)/token_data['CS'].copy().shift(29)
        token_data['CS variation 29-35 past days']=token_data['CS'].copy().diff(7).shift(29)/token_data['CS'].copy().shift(36)
        token_data['CS variation 36-42 past days']=token_data['CS'].copy().diff(7).shift(36)/token_data['CS'].copy().shift(43)
        token_data['CS variation 43-49 past days']=token_data['CS'].copy().diff(7).shift(43)/token_data['CS'].copy().shift(50)

        token_data['CS variation 2-7 following days']=token_data['CS'].copy().diff(6).shift(-6)/token_data['CS']
        token_data['CS variation 8-14 following days']=token_data['CS'].copy().diff(7).shift(-14)/token_data['CS'].copy().shift(-7)
        token_data['CS variation 15-21 following days']=token_data['CS'].copy().diff(7).shift(-21)/token_data['CS'].copy().shift(-14)
        token_data['CS variation 22-28 following days']=token_data['CS'].copy().diff(7).shift(-28)/token_data['CS'].copy().shift(-21)
        token_data['CS variation 29-35 following days']=token_data['CS'].copy().diff(7).shift(-35)/token_data['CS'].copy().shift(-28)
        token_data['CS variation 36-42 following days']=token_data['CS'].copy().diff(7).shift(-42)/token_data['CS'].copy().shift(-3)
       
        # control variables market conditions
        token_data['BTC price variation past 7 days'] = token_data['BTC_close'].copy().diff(7).shift(1)/token_data['BTC_close'].copy().shift(8)
        token_data['BTC price variation following 7 days'] = token_data['BTC_close'].copy().diff(7).shift(-6)/token_data['BTC_close'].copy().shift(1)
        token_data['BTC price variation past 31-60 days'] = token_data['BTC_close'].copy().diff(31).shift(30)/token_data['BTC_close'].copy().shift(61)
        token_data['BTC price variation past 30 days'] = token_data['BTC_close'].copy().diff(30).shift(1)/token_data['BTC_close'].copy().shift(31)
        token_data['BTC price variation following 30 days'] = token_data['BTC_close'].copy().diff(30).shift(-29)/token_data['BTC_close'].copy().shift(1)
        token_data['BTC price variation following 31-60 days'] = token_data['BTC_close'].copy().diff(30).shift(-59)/token_data['BTC_close'].copy().shift(-29)
                
        
        # Target variables 
            #daily variations     
        token_data['price variation unlock day'] = token_data['close'].copy().diff()/token_data['close'].copy().shift(1)
        token_data['price variation D-1'] = token_data['price variation unlock day'].copy().shift(1)
        token_data['price variation D-2'] = token_data['price variation unlock day'].copy().shift(2)
        token_data['price variation D-3'] = token_data['price variation unlock day'].copy().shift(3)
        token_data['price variation D-4'] = token_data['price variation unlock day'].copy().shift(4)
        token_data['price variation D-5'] = token_data['price variation unlock day'].copy().shift(5)
        token_data['price variation D-6'] = token_data['price variation unlock day'].copy().shift(6)
        token_data['price variation D-7'] = token_data['price variation unlock day'].copy().shift(7)
        token_data['price variation D-8'] = token_data['price variation unlock day'].copy().shift(8)
        token_data['price variation D+1'] = token_data['price variation unlock day'].copy().shift(-1)
        token_data['price variation D+2'] = token_data['price variation unlock day'].copy().shift(-2)
        token_data['price variation D+3'] = token_data['price variation unlock day'].copy().shift(-3)
        token_data['price variation D+4'] = token_data['price variation unlock day'].copy().shift(-4)
        token_data['price variation D+5'] = token_data['price variation unlock day'].copy().shift(-5)
        token_data['price variation D+6'] = token_data['price variation unlock day'].copy().shift(-6)
        token_data['price variation D+7'] = token_data['price variation unlock day'].copy().shift(-7)
        token_data['price variation D+8'] = token_data['price variation unlock day'].copy().shift(-8)
        token_data['price variation D+9'] = token_data['price variation unlock day'].copy().shift(-9)
        token_data['price variation D+10'] = token_data['price variation unlock day'].copy().shift(-10)

            #3 days period variations
        token_data['price variation following 3 days'] = token_data['close'].copy().diff(3).shift(-2)/token_data['close'].copy().shift(1)
        token_data['price variation following 4-6 days'] = token_data['close'].copy().diff(3).shift(-5)/token_data['close'].copy().shift(-2)
        token_data['price variation following 7-9 days'] = token_data['close'].copy().diff(3).shift(-8)/token_data['close'].copy().shift(-5)
        token_data['price variation following 10-12 days'] = token_data['close'].copy().diff(3).shift(-11)/token_data['close'].copy().shift(-8)
        token_data['price variation following 13-15 days'] = token_data['close'].copy().diff(3).shift(-14)/token_data['close'].copy().shift(-11)
       
        token_data['price variation past 3 days'] = token_data['close'].copy().diff(3).shift(1)/token_data['close'].copy().shift(4)
        token_data['price variation past 4-6 days'] = token_data['close'].copy().diff(3).shift(4)/token_data['close'].copy().shift(7)
        token_data['price variation past 7-9 days'] = token_data['close'].copy().diff(3).shift(7)/token_data['close'].copy().shift(10)
        token_data['price variation past 10-12 days'] = token_data['close'].copy().diff(3).shift(10)/token_data['close'].copy().shift(13)
        token_data['price variation past 13-15 days'] = token_data['close'].copy().diff(3).shift(13)/token_data['close'].copy().shift(16)

            #7 day periods variations        
        token_data['price variation following 7 days'] = token_data['close'].copy().diff(7).shift(-6)/token_data['close'].copy().shift(1)
        token_data['price variation following 8-14 days'] = token_data['close'].copy().diff(7).shift(-13)/token_data['close'].copy().shift(-6)
        token_data['price variation following 15-21 days'] = token_data['close'].copy().diff(7).shift(-20)/token_data['close'].copy().shift(-13)
        token_data['price variation following 22-28 days'] = token_data['close'].copy().diff(7).shift(-27)/token_data['close'].copy().shift(-20)

        token_data['price variation past 7 days'] = token_data['close'].copy().diff(7).shift(1)/token_data['close'].copy().shift(8)
        token_data['price variation past 8-14 days'] = token_data['close'].copy().diff(7).shift(8)/token_data['close'].copy().shift(15)
        token_data['price variation past 15-21 days'] = token_data['close'].copy().diff(7).shift(15)/token_data['close'].copy().shift(22)
        token_data['price variation past 22-28 days'] = token_data['close'].copy().diff(7).shift(22)/token_data['close'].copy().shift(29)

        final_token_data=token_data[(token_data['days since TGE']>0) 
                                    & (token_data['timestamp']>'2022-01-01') 
                                    & (token_data['close']>=0)]

    
        if final_dataset is None:
            final_dataset = final_token_data

        else:
            final_dataset = pd.concat([final_dataset, final_token_data], ignore_index=True)   
    
    
```

# Analysis

## Methodology

We conduct **multiple linear regression** analyses across different subsamples, with varying target and control variables, to assess the impact of token unlock events on token price variations.

### 1. **Explanatory Variable**

- **unlock (% of CS)**: This represents the percentage of circulating supply unlocked on a specific date. This is the explanatory variables in all our regressions.

### 2. **Target Variable**

- **Price Variations**:  
  The dependent (or target) variable in our regressions is always the percentage change in token prices. However, we examine price variations across different time horizons:
  
  - **Daily Variations VS Multi-Day Variations**: we analyze price changes on the specific day, but also over longer periods like 3-day and 7-day periods, to assess how unlock events influence price movements over longer windows.
  - **Lagged and Future Variations**: The regressions examine price changes both before and after the unlock, allowing us to determine whether the effects are anticipatory (before the unlock) or reactive (after the unlock).

### 3. **Control Variables**

To isolate the effect of the unlock event and prevent bias, we include several control variables in each regression:

- **CS variations on Past and Future Periods**:  
  This controls for any price variations due to adjacent unlock events. Without this control, the impact of surrounding unlocks could be mistakenly attributed to the event under analysis, skewing the results.
  
- **BTC Price Variations**:  
  Bitcoinâ€™s price fluctuations serve as a proxy for overall market conditions. As Bitcoin has a significant influence on altcoin prices, controlling for BTC movements helps ensure that price variations due to general market trends are not misattributed to token unlocks.

### 4. **Cluster Robust Standard Errors**

We employ **cluster robust standard errors** in our regressions. This method accounts for the error correlation within specific timestamps, as the price variations of different tokens are highly correlated within the same period. Clustering by timestamp allows us to control for this error correlation, providing more reliable estimates of the relationship between unlock events and price changes.


## I. Unlocks Greater Than 1% of Circulating Supply

Our first analysis focuses on unlock events where more than 1% of the circulating supply is released. this is an arbitrary threshold, that we chose because it is the same used by 6th men venture in their analysis, so we can compare results. These large unlock events are significant and are expected to be priced in by the market participants ahead of the unlock date. In contrast, smaller daily unlocks, which occur continuously, are less likely to be anticipated and reflected in the price in advance.

We focus on unlocks greater than 1% for the following reasons : 
* **Market Relevance**: Large unlocks (over 1%) are likely to capture market attention, prompting traders to adjust their positions, unlike the smaller, routine daily unlocks.
* **Reduced Bias**: These larger unlocks are less likely to be influenced by adjacent smaller unlocks. Smaller events occurring around the same time are typically insignificant, minimizing their effect on the price action related to larger unlocks.

The dataset includes 661 observations of unlock events where the percentage of circulating supply unlocked is greater than 1%.



```python
data = final_dataset[(final_dataset['unlock (% of CS)']>0.01)].dropna()
len(data)
```

## A. Daily price variations

We conducted a series of regression analyses to evaluate how large unlock events affect price variations at a daily level. We focused on price variations before, during, and after the unlock events to assess any systematic impact.

```python

# List of dependent variables (y)
dependent_vars = [
    'price variation D-8', 'price variation D-7', 'price variation D-6',
    'price variation D-5', 'price variation D-4', 'price variation D-3',
    'price variation D-2', 'price variation D-1', 
    'price variation unlock day', 
    'price variation D+1', 'price variation D+2', 'price variation D+3',
    'price variation D+4', 'price variation D+5', 'price variation D+6',
    'price variation D+7', 'price variation D+8'
]

# Independent variables (X) list
independent_vars = [
    'unlock (% of CS)', 'CS variation 36-42 past days', 'CS variation 29-35 past days', 'CS variation 22-28 past days',
    'CS variation 15-21 past days', 'CS variation 8-14 past days', 'CS variation 7 past days',  'CS variation 36-42 following days', 'CS variation 29-35 following days', 'CS variation 22-28 following days',
    'CS variation 15-21 past days', 'CS variation 8-14 past days', 'CS variation 2-7 following days',
    'BTC price variation past 7 days', 'BTC price variation following 7 days', 'BTC price variation past 30 days', 'BTC price variation past 31-60 days',
    'BTC price variation following 30 days', 'BTC price variation following 31-60 days'
]

results = {var: {'coefficients': {}, 'p-values': {}, 'conf_int': {}} for var in dependent_vars}

# Running regressions and storing results
for y_var in dependent_vars:
    if y_var in data.columns:
        X = data[independent_vars]
        X = sm.add_constant(X)
        
        y = data[y_var]
        
        # Fit the OLS model with cluster robust standard errors on 'timestamp'
        model = sm.OLS(y, X).fit(cov_type='cluster', cov_kwds={'groups': data['timestamp']})
        
        # Store coefficients, p-values, and confidence intervals
        for var in independent_vars:
            results[y_var]['coefficients'][var] = model.params[var]
            results[y_var]['p-values'][var] = model.pvalues[var]
            results[y_var]['conf_int'][var] = model.conf_int().loc[var]
    else:
        print(f"Warning: '{y_var}' not found in final_dataset columns.")

# Convert results to DataFrames 
coefficients_df = pd.DataFrame({y: results[y]['coefficients'] for y in dependent_vars if y in results}).T
p_values_df = pd.DataFrame({y: results[y]['p-values'] for y in dependent_vars if y in results}).T
conf_int_df = pd.DataFrame({y: results[y]['conf_int'] for y in dependent_vars if y in results}).T

# Plotting coefficients, p-values, and confidence intervals for each X variable
var = 'unlock (% of CS)' 
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plotting the coefficients on the primary y-axis
ax1.plot(coefficients_df.index, coefficients_df[var], marker='o', color='blue', label='Coefficient')
ax1.fill_between(
        coefficients_df.index,
        [conf_int_df[var][y][0] for y in dependent_vars],
        [conf_int_df[var][y][1] for y in dependent_vars],
        color='blue', alpha=0.3, label='95% Confidence Interval'
)
ax1.axhline(0, color='gray', linestyle='--')
ax1.set_ylabel('Coefficient', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

    # Plotting p-values on the secondary y-axis
ax2 = ax1.twinx()
ax2.plot(p_values_df.index, p_values_df[var], marker='x', color='red', label='P-Value')
ax2.axhline(0.05, color='red', linestyle='--', label='5% P-Value Threshold')
ax2.set_ylabel('P-Value', color='red')
ax2.tick_params(axis='y', labelcolor='red')

    # Combining legends from both axes
lines_1, labels_1 = ax1.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
ax1.legend(lines_1 + lines_2, labels_1 + labels_2, loc='upper right')

plt.title(f'Coefficient, P-Value, and Confidence Interval of {var}')
ax1.set_xlabel('Dependent Variable')
    
    # Set x-axis labels rotation for better readability
ax1.set_xticklabels(coefficients_df.index, rotation=45)

plt.grid(True)
plt.show()
```

## Intrepretation guide 

* **Blue Line (Coefficient)**: This represents the estimated coefficient for unlock (% of CS) on each price variation dependent variable.
    * A positive coefficient suggests that an increase in the unlock percentage leads to an increase in the corresponding price variation.
    * A negative coefficient indicates a decrease in price variation as unlock (% of CS) increases.
    
    
* **Shaded Area (95% Confidence Interval)**: This shows the range of likely values for the coefficient with 95% certainty. If this interval crosses zero, the effect of unlock (% of CS) on the price variation is not statistically significant for that variable.


* **Red Line with Crosses (P-Value)**: This represents the p-values for the coefficient of unlock (% of CS) on each dependent variable.
    * If the p-value is below the red dashed threshold line (0.05), the effect is statistically significant.
    * Higher p-values (closer to 1) indicate less evidence of a statistically significant effect.
    
    
## Results
From the analysis, we observe that most of the coefficients are not significantly different from zero, as their p-values exceed the 5% threshold, and their 95% confidence intervals overlap with zero. This indicates that for most time periods, unlock (% of CS) does not have a statistically significant effect on token price variations.

However, there are some noteworthy observations:

* Significant Impact on the 2nd Day Before the Unlock: The coefficient for the 2nd day before the unlock (D-2) is approximately -0.15 and statistically significant. This suggests that, on average, a 1% token unlock is associated with a 0.15% price drop two days before the unlock event.


* Other Potentially Significant Periods: There is also a significant negative coefficient on the 4th day after the unlock (D+4) and near-significant effects around the 3rd day after the unlock (D+3). These periods indicate potential downward pressure on prices leading up to and just after the unlock event.


* Day of Unlock and After: While there is a negative coefficient on the day of the unlock itself and the day before (D-1), and a slight positive coefficient the day after (D+1), these effects are not statistically significant.


Overall, the results suggest some price impact in the days leading up to and following the unlock event, but these effects are scattered and not consistently significant. To obtain clearer insights, we group the data into larger time windows, such as 3-day or 7-day periods, to smooth out the noise and better capture the overall effect of token unlocks on price variations. This approach should provide more interpretable results with less variability.

## Daily Unlocks

In this section, we focus on a completely different type of unlock event: the **daily unlocks**. We also include days with no unlocks, to give our model counter factuals where there is no unlock at all.


We isolate these daily unlocks by selecting events where the unlock amount is nearly identical to that of the **previous** and **following** days. By doing so, we aim to capture unlocks that occur regularly and consistently, excluding the "first day" of each daily unlock series to avoid introducing biases from any significant initial unlocks.

The reasoning behind this analysis is that these **daily unlocks** are typically expected to be **negligible** in terms of market impact. Because they happen on a regular basis with a very small percentage of the circulating supply unlocked each time, they are not likely to be seen as important or special events by market participants. Therefore, they are less likely to be anticipated or priced into the market ahead of time.

This subsample contains **52,730** observations, including **29,749 unlock events** for analysis, providing a robust dataset to explore the effects of tiny daily unlocks on token price variations.
